Bu notebook, 48 qubit'lik devrenin üçe bölünmüş halini (`P4_partA/B/C.qasm`) `Pennylane` kullanarak simüle eder.

```python
import pennylane as qml
from pennylane import numpy as np

def simulate_qasm(filename):
    dev = qml.device("default.qubit", wires=16, shots=1000)
    @qml.qnode(dev)
    def circuit():
        qml.from_qasm_file(filename)
        return qml.sample(wires=range(16))
    samples = circuit()
    bitstrings = ["".join(map(str, sample)) for sample in samples]
    print(f"\n📄 {filename} -- First 10 Samples:")
    for b in bitstrings[:10]:
        print(b)
```

```python
# ⬆️ Eğer bu dosyaları yüklemediysen, sol panele tıkla, sürükleyip bırak
simulate_qasm("P4_partA.qasm")
simulate_qasm("P4_partB.qasm")
simulate_qasm("P4_partC.qasm")
```
